```python
df = train_data.copy()
```

```python
df.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked","Initial"], axis = "columns", inplace = True)
```

```python
df
```

```python
inputs_features = df.drop("Survived", axis = "columns")
target = df.Survived
```

```python
dummies = pd.get_dummies(inputs_features.Sex)
dummies.head(3)
```

```python
inputs_features = pd.concat([inputs_features,dummies], axis = "columns")
inputs_features.head(3)
```

```python
inputs_features.drop(["Sex","male"], axis = "columns", inplace = True)
inputs_features.head()
```

```python
inputs_features.isna().sum()
```

```python
inputs_features.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(inputs_features, target, test_size = 0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_predict = model.predict(x_test)
y_predict
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import confusion_matrix

print("accuracy_score:",accuracy_score(y_test,y_predict))
print("f1_score:",f1_score(y_test, y_predict, average= None))
print("conusion Matrix:")
CM = confusion_matrix(y_test, y_predict)
print(CM)
```

# from confusion
* the first row represents the guesses for first class(i.e  survived)
* second row represents the 2nd class ( i.e did not survived)
* In first row we can see our model predicted 134 would survived and 134 survived , but it also said 34 survived, but in reality the did not survived. Which in short we could say. 134 correct guesses for Survived and 34 wrong guesses

* while in 2nd row we can see, our model predicted 66 did not survived, and it was was correct. but it also said 34 didnot survived but in reality they did survive. In short 66 correct guesses and 34 wrong guess for people who did not survive.
